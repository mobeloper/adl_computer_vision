```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data = pd.read_csv("Social_Network_Ads.csv")
```

```python
data.info()
```

```python
data.Purchased.unique()
```

```python
#Rules for Classification
# 1. Data must be complete
# 2. Data must be STRICTLY NUMERIC
# 3. Features and Label must be in the form on 2d np array
# 4. Scaling of features is OPTIONAL as per books but Prashant Nair will suggest to apply the same.
#           For Features:
#               StandardScaler(for normal data) or RobustScaler(if the data has outliers)
# 5. For Label
#      a. Binary Classification, label must be represented in the form of 0/1 (recommended to use pd.replace or LabelEncoder)
#      b. Multiclass Classification, label must be either DISCRETE NUMERICAL or DUMMY VARIABLE
```

```python
#Seperate data as features and Label
features = data.iloc[:,[0,1]].values
label = data.iloc[:,[2]].values
```

```python
#Feature Scaling
from sklearn.preprocessing import RobustScaler
rbFeatures = RobustScaler()
features = rbFeatures.fit_transform(features)
```

```python
#Label Transformation --- Not applicable as label is represented in the form of 0/1
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=10)
```

```python
#Architecting Model

model1 = tf.keras.Sequential()

model1.add(tf.keras.layers.Dense(units=6, activation="sigmoid", input_shape=(2,)))
model1.add(tf.keras.layers.Dense(units=6, activation="sigmoid"))
model1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
#Compile Model
#
# For Classification following loss functions mst be used:
# a. For Binary Classification: binary_crossentropy
# b. For Multiclass Classifiaction: categorical_crossentropy or sparse_categorical_crossentropy

model1.compile(optimizer='sgd', loss='binary_crossentropy', metrics='accuracy')
```

```python
#Fit the model
model1.fit(X_train,y_train, validation_data=(X_test,y_test), epochs=500)
```
